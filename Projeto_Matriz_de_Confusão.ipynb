```python
from sklearn.metrics import confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 10000 amostras, 6000 reais positivas (1), 4000 reais negativas (0)
y_true = np.array([1] * 6000 + [0] * 4000)

# 4500 verdadeiros positivos (TP), 1500 falsos negativos (FN),
# 3500 verdadeiros negativos (TN) e 500 falsos positivos (FP)
y_pred = np.array([1] * 4500 + [0] * 1500 + [0] * 3500 + [1] * 500)

# Calculando a matriz de confusão
cm = confusion_matrix(y_true, y_pred)

# Exibindo a matriz de confusão
print("Matriz de Confusão:")
print(cm)

# Visualizando a matriz de confusão
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Negativo', 'Positivo'], yticklabels=['Negativo', 'Positivo'])
plt.xlabel('Previsto')
plt.ylabel('Real')
plt.title('Matriz de Confusão')
plt.show()
```

```python
TN = cm[0, 0]
FP = cm[0, 1]
FN = cm[1, 0]
TP = cm[1, 1]

# Calculo as métricas
acuracia = (TP + TN) / (TP + TN + FP + FN)
precisao = TP / (TP + FP)
recall = TP / (TP + FN)
f1_score = 2 * (precisao * recall) / (precisao + recall)

# Exibindo os resultados
print(f"Acurácia: {acuracia:.4f}")
print(f"Precisão: {precisao:.4f}")
print(f"Recall (Revocação): {recall:.4f}")
print(f"F1-Score: {f1_score:.4f}")
```
